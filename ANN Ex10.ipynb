```python
#C-61
import tensorflow as tf
import numpy as np
np.random.seed(0)
x = np.random.rand(1000, 2)
y = np.random.randint(2, size=1000)
x_train, x_test = x[:800], x[800:]
y_train, y_test = y[:800], y[800:]
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid', input_dim=2)
])
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=False),
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10)
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
import numpy as np
(x_train, y_train), _ = mnist.load_data()
idx = np.random.randint(0, len(x_train))
digit_image = x_train[idx]
digit_label = y_train[idx]
plt.figure(figsize=(4, 4))                # Bigger figure size for clarity
plt.imshow(digit_image, cmap='gray_r', interpolation='nearest')  # Better colormap & rendering
plt.title(f"Digit: {digit_label}", fontsize=18)
plt.axis('off')
plt.tight_layout()
plt.show()
```
